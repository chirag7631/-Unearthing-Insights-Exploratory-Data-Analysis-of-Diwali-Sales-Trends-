```python
replacement_mapping = { 
"Men":"M",
"Women":"W"}
```

```python
df['Gender'] = df['Gender'].replace(replacement_mapping)
```

```python
df.head(34)
```

## performing EDA


```python
sx = sns.countplot(x = 'Gender',data=df)
for bars in sx.containers:
    sx.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender',y='Amount',data = sales_gen)
```

*From above chart we can conclude that most of buyers are Women and even the purchasing power of women are greater than men*

```python
sx = sns.countplot(data=df,x = 'Age group',hue = 'Gender')
for bars in sx.containers:
    sx.bar_label(bars)
```

```python
sales_gen = df.groupby(['Age group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Age group',y='Amount',data = sales_gen)
```

*From above graphs we can conclude that most of the buyers are of age group between Adults women*


```python
sales_gen = df.groupby(['ship-state'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'ship-state',y='Amount',data = sales_gen)
```

*from above graphs we can see that the most of orders & total sales/amount are from  maharashtra, karnataka and uttar pradeshrespecrively*
